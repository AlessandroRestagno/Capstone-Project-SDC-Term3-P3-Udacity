```python
import cv2
import os
import glob

recindex = 6

vidcap = cv2.VideoCapture(f'D:/Projects/SDC/Term3/Capstone-Project-SDC-Term3-P3-Udacity/recs/videoplayback{recindex}.mp4')
success,image = vidcap.read()
count = 0
jpg_folder = f"D:/Projects/SDC/Term3/Capstone-Project-SDC-Term3-P3-Udacity/recs/{recindex}/"

if not os.path.exists(jpg_folder):jpg_folder = f"D:/Projects/SDC/Term3/Capstone-Project-SDC-Term3-P3-Udacity/recs/{recindex}/"
    os.makedirs(jpg_folder)
else:
    output_file_list = glob.glob(os.path.join(jpg_folder, "*.*"))
    for f in output_file_list:
        os.remove(f)

while success:
  image = cv2.resize(image,(800, 600), interpolation = cv2.INTER_CUBIC)  
  cv2.imwrite(jpg_folder + f'frame{count}-{recindex}.jpg', image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```
